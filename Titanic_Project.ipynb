## Handling missing values

```python
df['age'].fillna(df['age'].mean(), inplace=True)

df['embarked'].fillna(df['embarked'].mode()[0], inplace=True)
```

## Deal with duplicates if exist.

```python
df.drop_duplicates(inplace=True)
```

## Save your modified dataframe.

```python
df.to_csv('titanic_cleaned.csv', index=False)
```

## Showing how many passengers survived and how many did not

```python
import matplotlib.pyplot as plt

survival_counts = df['survived'].value_counts()

plt.figure(figsize=(6,4))
plt.bar(['Did not survive', 'Survived'], survival_counts, color=['red', 'green'])
plt.title('Survival Count of Titanic Passengers')
plt.ylabel('Number of Passengers')
plt.show()
```

## Show the percentage of passengers in each class (Pclass).

```python
class_counts = df['pclass'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(class_counts, labels=['1st Class','2nd Class','3rd Class'], 
        autopct='%1.1f%%', startangle=90, colors=['gold', 'silver', 'lightblue'])
plt.title('Passenger Distribution by Class')
plt.show()
```

## Showing the number of passengers by embarkation port (Embarked).

```python
port_counts = df['embarked'].value_counts()

plt.figure(figsize=(6,4))
plt.bar(port_counts.index, port_counts.values, color='skyblue')
plt.xlabel('Port')
plt.ylabel('Number of Passengers')
plt.title('Number of Passengers by Embarkation Port')
plt.show()
```

## Fare vs Age 

```python
colors = df['survived'].map({0:'red', 1:'green'})  # لون حسب الناجين

plt.figure(figsize=(8,6))
plt.scatter(df['age'], df['fare'], c=colors, alpha=0.6)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Fare vs Age Colored by Survival')
plt.show()
```

## Survival Rate by Gender

```python
gender_survival = df.groupby('sex')['survived'].mean() * 100

plt.figure(figsize=(6,4))
plt.bar(gender_survival.index, gender_survival.values, color=['blue','pink'])
plt.ylabel('Survival Rate (%)')
plt.title('Survival Rate by Gender')
plt.ylim(0,100)
plt.show()
```
